```python
k = 1.5
DC_dict = {
    'pay_func': lambda X: F.relu(torch.sum(torch.abs(X), dim = 1, keepdims = True) - k),
    'lb1': -5.,
    'ub1': 5.,
    'lb2': -2.,
    'ub2': 2.,
}
payoff_dict = {
               'DC': DC_dict,
               }
```

```python
import numpy as np
import pandas as pd
import torch
from scipy.stats import random_correlation
import torch.nn.functional as F
import matplotlib.pyplot as plt
import seaborn as sns
#%matplotlib notebook

import torch
import torch.nn as nn
import torch.nn.functional as F
import math
from tool import *
from data_generator import DataGenerator
device = 'cuda:1'
```

```python
class NN(nn.Module):
    def __init__(self, input_dim, init_basket, k, std_X):
        super(NN, self).__init__()
        self.name = 'NN'
        self.units = init_basket.shape[1]
        init_basket = torch.as_tensor(init_basket, dtype=torch.float32)

        self.w = nn.Parameter(init_basket)
        nn.init.normal_(self.w)
        
        self.v = nn.Parameter(
            torch.normal(mean=0.,
                         std=math.sqrt(2 / (1 + self.units)),
                         size=(self.units, 1)))
        #nn.init.xavier_normal_(self.v)
        self.activation = F.relu
        self.k = k
        self.std_X = std_X
    def forward(self, x):
       
        x = self.activation(torch.matmul(x, self.w) - self.k/self.std_X)
        
        x = torch.matmul(x, self.v)
        return x
```

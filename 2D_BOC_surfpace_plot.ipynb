```python
import numpy as np
import pandas as pd
import pickle
import matplotlib.pyplot as plt
%matplotlib notebook
import matplotlib.tri as mtri
from mpl_toolkits.mplot3d import Axes3D
```

```python
data = pd.read_pickle('./result/report_dim2.pickle')
```

```python
y_test = data['BOC']['y_test']
pred = data['BOC']['NN_pred']
x1 = data['BOC']['X_test'][:,0]
x2 = data['BOC']['X_test'][:,1]
X_train = data['BOC']['X_train']
y_train = data['BOC']['y_train']
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(1, 1, 1, projection='3d')
ax.scatter(x1[0:1000],
           x2[0:1000],
           pred[0:1000].ravel(),
           marker='.',
           s=15,
           c="red",
           alpha=0.4,  )
ax.view_init(elev = 26, azim = 49)
ax.set_xlabel(r'$x_1$', rotation=150,fontsize=13)
ax.set_ylabel(r'$x_2$',fontsize=13)
ax.set_zlabel('Payoff value', rotation=60,fontsize=13)
plt.rc('xtick',labelsize=10)
plt.rc('ytick',labelsize=10)
triang = mtri.Triangulation(X_train[:,0], X_train[:,1])

trisurf = ax.plot_trisurf(triang,
                          y_train.ravel(),
                          linewidth=0.,
                          antialiased=False,
                          edgecolor='none',
                          cmap=plt.get_cmap('Blues'))
fig.colorbar(trisurf, ax=ax, shrink=0.7, aspect=10, cmap=plt.get_cmap('Blues'), location = 'left')

plt.show()
plt.savefig('boc_3d_1.png')
#fig.savefig("bbox_inches_tight.pdf", bbox_inches = None)
```
